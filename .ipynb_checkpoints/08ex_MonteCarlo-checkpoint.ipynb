1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
%matplotlib inline
```

```python
T=3.052
Tl=np.array([1000])
Pb=np.array([0])
t=np.array([0])

def p(t):
    return 1-2**(-t/T)


while Tl[-1]!=0:
    count=0
    particles=np.random.random(Tl[-1])
    for particle in particles:
        if particle<p(t[-1]): count+=1
    Tl=np.append(Tl,Tl[-1]-count)
    Pb=np.append(Pb,Pb[-1]+count)
    t=np.append(t,t[-1]+(1/60))
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(14,7))
ax1.set_title('Tallium',fontsize=20)
ax1.set_xlabel('time (minutes)',size=14)
ax1.set_ylabel('number of atoms',size=14)
ax1.plot(t,Tl)

ax2.set_title('Lead',fontsize=20)
ax2.set_xlabel('time (minutes)',size=14)
ax2.set_ylabel('number atoms',size=14)
ax2.plot(t,Pb,color='orange')
plt.show()
```

```python
def CDF(p):
    return -T*np.log2(1-p)
p=np.random.random(1000)
t=CDF(p)
t=np.sort(t)
decays,bins=np.histogram(t,30,(0,30))
Tallium=np.zeros(30)
Tallium[0]=int(1000-decays[0])
for i in range (1,30):
    Tallium[i]=Tallium[i-1]-decays[i]
decays=list(decays)
bin_centers=(bins[0:-1]+bins[1:])*0.5
Tallium=list(Tallium)
for i in range(len(Tallium)):
    Tallium[i]=int(Tallium[i])

plt.figure(figsize=(12,7))
plt.step(bin_centers,Tallium,where='mid',linewidth=2,label='Tallium')
plt.axvline(3.052,linestyle='--',color='orange',label= r'$\tau$')
plt.legend(loc='best',prop={'size':14})
plt.xlabel('time (minutes)',size=14)
plt.ylabel('number of atoms',size=14)
plt.title('Tallium decay vs time',fontsize=20)
plt.show()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
#setting the condition on b
E=7.7#Mev
Z=79
e=1.602*10**(-19)
epsilon0=8.8544*10**-12
b=(Z*e**2)/(2*np.pi*epsilon0*E*1.602*10**(-13))

a0=5.291777721092*10**(-11)
#Using the Box-Muller method to produce one million particles distributed as a Gaussian
def bm(z,sigma=a0/100):
    return np.sqrt(-2*(sigma**2)*np.log(1-z))

z=np.random.random(1000000)
r=bm(z)
bounce_back=r[r<b]
print('The fraction of particles whose scattering angle is grater than pi/2 is:{:.4f}%'.format(len(bounce_back)*100/1000000))
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return np.power(np.sin(1/(x*(2-x))),2)
N=100000
count=0
fx=f(2*np.random.random(N))
y=np.random.random(N)
for i in range(N):
    if y[i]<fx[i]: count+=1
I=2*count/N
E=2.*fx.std()/np.sqrt(N)
print('I=',I,'Error=',E)
```

```python
x=2*np.random.random(N)
func=f(x)
I=2*func.mean()
E=2.*func.std()/np.sqrt(N)
print('I=',I,'Error=',E)
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
N=1000000
random=np.random.rand(N,2)
samples=2*random-1
norm=np.sqrt((samples[:,0]**2)+(samples[:,1]**2))
I=4*len(norm[norm<=1])/N
print(I)
```

```python
from scipy import linalg as la
random=np.random.rand(N,10)
samples=2*random-1
norm=np.zeros(len(samples))
for i in range (N):
    norm[i]=la.norm(samples[i])
I=(2**10)*len(norm[norm<=1])/N
print(I)
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
#Defining integrand function
def f(x):
    return x**(-0.5)/(1+np.exp(x))

#Defining the weight function
def w(x):
    return 0.5/(x**0.5)

n=1000
random=np.random.rand(n)
sample=np.power(random,2)
I=(1./n)*np.sum(f(sample)/w(sample))
print(I)
```
