2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
#PARAMETRI
N_alpha=1000000
E=7.7e6*1.602176634e-19 #J
Z=79
elett=-1.602e-19#C
eps_0=8.85418781762e-12 #F/m
a_0=5.291777721092e-11 #metri

def theta(b):
    return 2*np.arctan(  (Z*elett**2) /(2*np.pi*eps_0*E*b) )
#ESTRAZIONI RANDOM 
x = np.random.normal(0,a_0/100,N_alpha)# fascio particelle
y = np.random.normal(0,a_0/100,N_alpha)#sarà l'altezza y della singola particella alpha

counter=0
for i in range(N_alpha):
    if (np.abs(x[i])<a_0/1000):#sono nei pressi del bersaglio lungo l'asse x
        if theta(y[i])> np.pi/2:
            counter=counter+1
print("fraction of particles bounced back",(counter*100)/N_alpha,"%")
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return (np.sin(1/(x*(2-x))))**2

x=np.linspace(0.001,1.999,1000)
plt.plot(x,f(x),'r-')

#"true" value of the integral:
func = lambda x: (np.sin(1/(x*(2-x))))**2
res, err = quad(func, 0, 2)
print("'true' value of the integral (from scipy.integrate.quad):",np.round(res,3),"( error",np.round( err,3),")")

# hit/miss method
N=100000
count=0
for i in range(N):
    x=2*np.random.random()
    y=np.random.random()
    if y<f(x): count+=1
I=2*count/N
err_hit_miss= np.abs(I-res)/res #errore hit/miss method
print("Value of the integral (from hit/miss method):",np.round( I,3),"( error",np.round( err_hit_miss,3),")")

#the mean value method
count=0
for i in range(N):
    x=2*np.random.random()
    count=count+f(x)
I=2*(count/N)
err_mean_v= np.abs(I-res)/res #errore mean value method
print("Value of the integral (from the mean value method):",np.round( I,3),"( error",np.round( err_mean_v,3),")")
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
def f(x,y):
    return np.power(x,2)+np.power(y,2)
N=10000000
count=0
b=1
a=-1
epsilon=0.00001
for i in range(N):
    x=(b - a) * np.random.random() + a
    y=(b - a) * np.random.random() + a
    if (f(x,y)<1+epsilon):
        count=count+1 
I=4*count/N
print("The area of a circle of unit radius:",I)
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
def f(x):
    return np.power(x,-1/2)/(np.exp(x)+1)
def w(x):
    return np.power(x,-1/2)

M=10000
x=np.linspace(0.001,1.0,M)
plt.plot(x,w(x),'r-', label="w(x)")
plt.plot(x,f(x),'b-', label ="f(x)")
plt.legend()
plt.show
# Monte Carlo integration 
count=0
for i in range(M):
    x=np.random.random()
    y=(w(0.001))*np.random.random()
    if y<w(x): 
        count=count+1
I=w(0.001)*count/M
#print(I)#è circa 2

cont=0
x=np.linspace(0.001,0.999,M)
for i in range(M):
    cont=cont+(f(x[i])/w(x[i]))
print(np.round(((cont*I)/M),2))
print("(the obtained value is slightly underestimated but I'm not able to improve it)")
```
