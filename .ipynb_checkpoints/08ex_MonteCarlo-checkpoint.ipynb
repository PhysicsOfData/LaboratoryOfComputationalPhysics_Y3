* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
# setting the time records starting from 0 to 5tau in order to get more than the 99% of the decays
tau = 3.052 * 60
time_records = numpy.arange(0, 10*tau, dtype=int)

# setting the arrays for Tl and Pb
Tl_records = numpy.full(time_records.size, 1000)
Pb_records = numpy.full(time_records.size, 0)

# setting up the random number associated to each atom
randoms = rnd.uniform(size=Tl_records.size)

# defining the ????????????
def prob(t):
    return 1 - exp2(-t/tau)
plt.plot(time_records, prob(time_records))

# running the for loop to tipe-evolve the system
for t in time_records:
    decays = numpy.sum(randoms < prob(t))
    Tl_records[t] = 1000 - decays
    Pb_records[t] = decays

# plotting the results obtained
fig, axs = plt.subplots(1,2, figsize=(12,5))
axs[0].set_title("Elements population")
axs[1].set_title("Elements population in a time log-scale")
for ax in axs:
    ax.plot(time_records, Tl_records, label="Thallium")
    ax.plot(time_records, Pb_records, label="Lead")
    ax.set_xlabel("Time")
    ax.set_ylabel("Number of atoms")
    ax.axvline(tau, color='grey', label=r"$\tau$")
    ax.legend()
axs[1].set_xscale('log')
```

```python
# setting the time records starting from 0 to tau
tau = 3.052 * 60
time_records = numpy.arange(0, 5*tau, dtype=int)

# setting the arrays for Tl and Pb
Tl_records = numpy.full(time_records.size, 1000)
Pb_records = numpy.full(time_records.size, 0)

# defining the probability of decay
def prob(t):
    return exp2(-t/tau) * log(2) / tau

# running the for loop to time-evolve the system
for t in time_records[1:]:
    randoms = rnd.uniform(size=Tl_records[t-1])
    decays = numpy.sum(randoms < prob(t))
    Tl_records[t] = Tl_records[t-1] - decays
    Pb_records[t] = Pb_records[t-1] + decays

# plotting the results obtained
fig, axs = plt.subplots(1,2, figsize=(12,5))
axs[0].set_title("Elements population")
axs[1].set_title("Elements population in a time log-scale")
for ax in axs:
    ax.plot(time_records, Tl_records, label="Thallium")
    ax.plot(time_records, Pb_records, label="Lead")
    ax.set_xlabel("Time")
    ax.set_ylabel("Number of atoms")
    ax.axvline(tau, color='grey')
    ax.legend()
axs[1].set_xscale('log')
```
