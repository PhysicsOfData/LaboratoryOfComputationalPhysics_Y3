## Exercise 1

Create a vectorized version of log and exp math function for 1D array A = [2, 5, 10, 3, 8]

Results should be: 
+ [0.6931472 1.609438  2.3025851 1.0986123 2.0794415]
+ [7.3890562e+00 1.4841316e+02 2.2026465e+04 2.0085537e+01 2.9809580e+03]

````python
import os
import math as m
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
os.environ["NUMBA_ENABLE_CUDASIM"] = "1"
from numba import jit, njit, cuda, vectorize, float32, int32, float64, int16
````

````python
@vectorize([float32(int32)])
def vec_log(x):
    return m.log(x)

@vectorize([float32(int32)])
def vec_exp(x):
    return m.exp(x)
````

````python
a=np.array([2,5,10,3,8], dtype='int32')

result1=vec_log(a)
result2=vec_exp(a)
print('Vectorized version of log:',result1)
print('Vectorized version of exp:',result2)
````

## Exerice 2
Compute the value of a Gaussian probability density function at $x$ with $mean = 1$, $\sigma = 1$, lower and upper bound in $(-3, 3)$ and $size = 100000$

````python
x=np.linspace(-3,3,100000)
gauss=stats.norm.pdf(x,loc=1,scale=1)
plt.figure()
plt.plot(x,gauss)
plt.show()
print('Value of a Gaussian proabability density function at x with mean=1, sigma=1:',gauss)
````

````python
@vectorize([float64(float64,int32,int32)])
def vec_gauss(x,mean,std):
    y=(x-mean)/std
    value=m.exp(-y*y/2)/m.sqrt(2*m.pi)
    return value

result=vec_gauss(x,1,1)
print(np.allclose(gauss,result))
plt.figure()
plt.plot(x,result)
plt.show()
print('Value of a Gaussian proabability density function at x with mean=1, sigma=1:',result)
````

## Exercise 3

Create a "zero suppression" function. A common operation when working with waveforms is to force all samples values below a certain absolute magnitude to be zero, as a way to eliminate low amplitude noise. 
Plot the data before and after the application of the zero_suppress function.

$thresold = 15$

````python
%matplotlib inline
from matplotlib import pyplot as plt

n = 100000
noise = np.random.normal(size=n) * 3
pulses = np.maximum(np.sin(np.arange(n) / (n / 23)) - 0.3, 0.0)
data = ((pulses * 300) + noise).astype(np.int16)

#put your code here

@vectorize([float32(int16,int16)])
def zero_suppression(values, threshold):
    if np.abs(values) >= threshold:
        return values
    else:
        return 0.0

zero_suppressed=zero_suppression(data,15)
x=np.arange(100000)

figure,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(14,5))

ax1.plot(x,data)
ax1.set_title('Data',fontsize=18)

ax2.plot(x,zero_suppressed,color='orange')
ax2.set_title('Zero suppressed',fontsize=18)
plt.show()
````

## Exercise 4

Calculate the Sigmoid kernel between the matrix X and Y defined as below. The Sigmoid kernel is defined as:

$k(x,y) = \tanh(\alpha x^T y + c) $

````python
X = np.random.rand(3,3)
Y = np.random.rand(3,3)
alpha=2.
c=5.
````

````python
@jit(nopython=True)
def Sigmoid_kernel(X,Y,alpha,c):
    return np.tanh(alpha*(X.T*Y)+c)

Sigmoid=Sigmoid_kernel(X,Y,alpha,c)
print(Sigmoid)
````

## Exercise 5

Create a kernel function similar to the ```double_kernel``` see during the lecture and create a new function that takes a 3 dimensional matrix as input calculating the $cos$ for each element and then returns the result. The shape of the matrix must be $256X256X256$. The matrix can be randomly generated

````python
@cuda.jit
def triple_kernel_3D(io_array):
    x,y,z=cuda.grid(3)
    if x<io_array.shape[0] and y<io_array.shape[1] and z<io_array.shape[2]:
        io_array[x,y,z]=m.cos(io_array[x,y,z]) #do the computation

data=np.random.rand(256,256,256)*2*np.pi

# Configure the blocks
threadsperblock = (8, 8, 4)
blockspergrid_x = int(m.ceil(data.shape[0] / threadsperblock[0]))
blockspergrid_y = int(m.ceil(data.shape[1] / threadsperblock[1]))
blockspergrid_z = int(m.ceil(data.shape[2] / threadsperblock[2]))
blockspergrid = (blockspergrid_x, blockspergrid_y,blockspergrid_z)

triple_kernel_3D[blockspergrid, threadsperblock](data)
print(data)
````

## Exercise 6

Create a matrix multiplication kernel function, called ```matmul``` that takes as input two 2D matrices:
+ A of shape $24x12$
+ B of shape $12x22$
and that computes the multiplication and put the results into a third matrix C of shape $24x12$

A and B must be randomly generated and only int values are allowed.


````python
from __future__ import division
from numba import cuda
import numpy
import math

# complete the code
@cuda.jit
def matmul(A, B, C):
    """Perform matrix multiplication of C = A * B
    """
    row, col = cuda.grid(2)
    if row < C.shape[0] and col < C.shape[1]:
        tmp = 0.
        for k in range(A.shape[1]):
            tmp += A[row, k] * B[k, col]
        C[row, col] = tmp

# Initialize the data arrays
A = np.random.randint(1,5,(24,12))
B = np.random.randint(1,5,(12,22))
C = np.zeros((24,12),int)

# Configure the blocks
threadsperblock = (8,8)
blockspergrid_x = int(m.ceil(A.shape[0] / threadsperblock[0]))
blockspergrid_y = int(m.ceil(B.shape[1] / threadsperblock[1]))
blockspergrid = (blockspergrid_x, blockspergrid_y)

matmul[blockspergrid, threadsperblock](A,B,C)
print(C)
````
