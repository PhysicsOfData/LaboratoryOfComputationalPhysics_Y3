1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
f = open("simple_data.txt", "w")

np.random.seed(8493) 
random_vector = np.random.rand(25)

for i in range (0,25,1):
    f.write(str(random_vector[i]))
    f.write('\n')

f.close()
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
M = random_vector.reshape((5,5))

f = open("data.txt", "w")

for i in range (0,5,1):
    for j in range (0,5,1):
        f.write(str(M[i][j]))
        f.write(" ")
    f.write('\n')
    
f.close()
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
M = []

with open("data.txt" , mode='r') as f:
    for line in f:
        M.append(line)
        
writer = csv.writer(open("data.csv", 'w'))
writer.writerows(M)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
def converter(line,n_bit,n_padding):
    #pop out non-number elements 
    line.pop(0)
    line.pop(0)
    #pop out non-number elements 
    for i in range(0,n_padding,1):
        line.pop()
        
    size = len(line)
    n_char = size/n_bit
    
    card_number = ''
    
    for i in range(0,int(n_char),1):
        number = 0
        for n in range(0,n_bit,1):                          #6 bit binary conversion
            number = number + int(line[i*n_bit+n])*2**(5-n)
        card_number += chr(number)                          #char conversion
    return card_number


lines = []

with  open('credit_card.dat','rb') as file:
    for line in file:
        lines.append(list(str(line)))
        
for i in lines:
    print(converter(i,6,7))
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
data = json.load(open('user_data.json'))
writer = csv.writer(open("user_data.csv", 'w'))

header = data[0].keys()
writer.writerow(header)

for i in data:
    if (i['CreditCardType']=="American Express"):
        writer.writerow(i.values())

```
