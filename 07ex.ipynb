```python
import numpy as np
from scipy import linalg as la
from matplotlib import pyplot as plt
x = np.random.normal(0,1,1000)
x2=np.random.normal(0,3,1000)+x
x3=2*x+x2
A=np.array([(x),(x2),(x3)])

C=np.cov(A)
l, V = la.eig(C)
print (l,'\n')
print (V,'\n')

U, spectrum, Vt = la.svd(A)
print (spectrum,'\n')

print (U,'\n')
print (Vt,'\n')
```

```python
plt.figure(figsize=(10,10))
# the original data distribution

plt.scatter(A[0,:], A[1,:], alpha=0.6)
scale_factor=0.1
h,k = la.eig(np.cov(A))
for li, vi in zip(h, k.T):
    print (li, vi)
    # the line is defined by means of its beginning and its end 
    plt.plot([2, scale_factor*li*vi[0]], [0, scale_factor*li*vi[1]],[0, scale_factor*li*vi[2]], 'r-', lw=3)
plt.axis([-9,9,-9,9])
print k
print h
```

```python
Lambda=np.diag(h)
print (Lambda)
print ("A.trace():", np.cov(A).trace())
print ("Lambda.trace():", Lambda.trace())

print (Lambda[0,0]/Lambda.trace())
```
