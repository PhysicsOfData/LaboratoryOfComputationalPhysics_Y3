```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 3.052*60
N_Tl = 1000
N_Pb = 0
```

```python
def p1(t, tau=3.052*60):
    return 1 - 2**((-t)/tau)
Tl_decay = []
for i in range(int(tau)):
    Tl_decay.append(p1(i))
Tl_decay_1000 = N_Tl - np.array(Tl_decay)*N_Tl
Pb_add = np.array(Tl_decay)*1000
time = np.arange(int(tau))
plt.plot(time, Tl_decay_1000, label = "Tl decay")
plt.plot(time, Pb_add, label = "Pb add")
plt.title("Tl208  decays to  Pb208")
plt.xlabel("Time (s)")
plt.ylabel("Number of atom")
plt.legend(loc = "best")
plt.show()
```

```python
import numpy as np
def f_circle(x, y):
    return 1 if (x*x + y*y) <= 1 else 0
N = 100000
count = np.zeros(100)
for i in range(100):
    x = 2*np.random.random(N) -1
    y = 2*np.random.random(N) -1
    for j in range(N):
        count[i] += f_circle(x[j],y[j])
I = 4*count/N
print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

```python
def f(x):
    return (x**(-0.5))/(np.exp(x)+1)
def w(x):
    return np.sqrt(x)**-1
def inv_trans(x):
    return np.square(x)/4
```
