2\. Convert the following function into a pure function with no global variables or side effects

````python
x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

````python
#Answer
def f(alist, x=5):
  blist=alist.copy()
  for i in range(x):
         blist.append(i)
  return blist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist)
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello(func):
    def wrapper(*args, **kwargs):
        print('Hello!')
        func(*args, **kwargs)
    return wrapper

@hello
def square(x):
    return x*x

square(2)
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
#a)
def factorial(x):
  if x==1: return x
  return x*factorial(x-1)

print(factorial(5))

#b)
def factorial(x):
  t=1
  for i in range (1, x+1):
    t=t*i
  return t

print(factorial(5))
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
import math as m
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
lambdas=[lambda r: (2 * m.pi * r,m.pi*r**2,4/3*m.pi*r**3) for i in range(len(radii))]
for element in densities.keys():
  for r,func in zip(radii,lambdas):
    print("Element: ",element,", Radius: ",r)
    print("Weights: ", [x*y for x,y in zip(func(r),densities[element])])
````
